# Modelo de regressão linear



```julia
# Preparando os dados
c = ones(size(Rme,1))
x = [c Rme]
y = copy(Re)
```

```julia
# Calculo dos coeficientes
b2 = inv(x'x)*x'y
b =x\y
u = y - x*b
```

```julia
# Coeficiente de determinação para avaliar o modelo
RQuadrado = 1 - var(u)/var(y)
```

```julia
println("Resultado da regressão")
println("R²: $RQuadrado")
println("Número de observações: ", size(Re,1))
```

```julia
xTicksLoc = [Date(1980), Date(1990), Date(2000), Date(2010)]

xTicksLab = Dates.format.(xTicksLoc, "Y")

p1 = plot( dn,
           Rme,
           linecolor = :green,
           legend = false,
           xticks = (xTicksLoc, xTicksLab),
           ylim = (-25,25),
           title = "Time Series Plot - Taxa de Retorno do Mercado ao Longo do Tempo",
           titlefont = font(10),
           ylabel = "%" )
display(p1)
```

```julia
# Scatter Plot
p1 = scatter( Rme,
              Re,
              fillcolor = :green,
              legend = false,
              xlim = (-40,40),
              ylim = (-40,60),
              title = "Scatter Plot: Taxa de Retorno do Mercado x Retorno Real",
              titlefont = font(10),
              xlabel = "Taxa de Retorno do Mercado, %",
              ylabel = "Retorno Real, %")
plot!([-40;60], [-40;60], color = :black, linewidth = 0.5)   
display(p1)
```

```julia
# Histograma
xGrid = -25:0.1:15
pdfX  = pdf.(Normal(mean(Rme), std(Rme)), xGrid) 

p1 = histogram( Rme,
                bins = -25:1:15,
                normalized = true,     
                label = "Histograma",
                title = "Histograma da Taxa de Retorno do Mercado",
                titlefont = font(10),
                xlabel = "Taxa de Retorno do Mercado, %",
                legend = :topleft)  
plot!(xGrid, pdfX, linewidth = 3, label = "fitted N()")
display(p1)
```
